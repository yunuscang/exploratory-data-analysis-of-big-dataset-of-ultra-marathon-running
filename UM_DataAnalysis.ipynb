```python
df1["age_group"]=df1["athl_age"].apply(lambda x: "Under 18" if x < 18 else
                                                         "18-25" if 18 <= x < 25 else
                                                         "25-35" if 25 <= x < 35 else
                                                         "35-45" if 35 <= x < 45 else
                                                         "45-55" if 45 <= x < 55 else
                                                         "+55")
```

```python
#Fixing the dates
```

```python
df1[df1["race_day"].str.contains(r"\.\.")]
```

```python
df1.loc[df1["race_day"].str.contains(r"\.\.", na=False), "race_day"] = df1["race_day"].str[4:]
```

```python
df1["race_day"].value_counts()
```

```python
df1.loc[df1["race_day"].str.contains(r"\.\-", na=False), "race_day"]=df1["race_day"].str[4:]
```

```python
df1["race_day"].value_counts()
```

```python
df1["race_month"]=df1["race_day"].str.split(".").str.get(1).astype(int)
```

```python
df1["race_season"]=df1["race_month"].apply(lambda x: "Spring" if 2<x<=5 else
                                                     "Summer" if 5<x<=8 else
                                                     "Fall"   if 8<x<=11 else
                                                     "Winter")
```

```python
df1.drop("race_month",axis=1,inplace=True)
```

```python
df1.groupby(["race_season","age_group"])["athl_avg_speed"].mean()
```

```python
df1.groupby("race_season")["athl_avg_speed"].agg(["mean","count"]).sort_values("mean",ascending=False)
```
